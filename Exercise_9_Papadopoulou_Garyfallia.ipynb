#Question 2

The **reset()** function is used to initialize the environment at the beginning of an episode. It resets the environment to its initial state and returns the initial observation, which represents the starting state of the agent.

The **step(action)** function applies an action to the environment and advances it by one time step. It returns the next observation, the reward received after performing the action, and information indicating whether the episode has ended either because the task was completed or because a time or step limit was reached.

The **render()** function is used to visualize the environment. It allows the user to observe the current state of the environment and is mainly used for debugging and demonstration purposes.

#Question 3

The **observation space** consists of an 8-dimensional continuous vector that represents the state of the lander, including its position, velocity, angle, angular velocity, and whether each landing leg is in contact with the ground.

The **action space** is discrete and includes four possible actions: doing nothing, firing the left orientation engine, firing the main engine, or firing the right orientation engine.

The **reward function** is designed to encourage a safe and efficient landing. The agent receives positive rewards for approaching the landing pad, reducing speed, and landing successfully, while negative rewards are given for crashing, excessive fuel consumption, or moving away from the target.

#Question 5

```python
import numpy as np

env = gym.make("LunarLander-v3")

num_episodes = 5
episode_rewards = []

for episode in range(num_episodes):
    observation, info = env.reset()
    total_reward = 0

    while True:
        action = env.action_space.sample()
        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

        if terminated or truncated:
            break

    episode_rewards.append(total_reward)
    print(f"Episode {episode + 1} reward: {total_reward:.2f}")

average_reward = np.mean(episode_rewards)
print(f"\nAverage reward over {num_episodes} episodes: {average_reward:.2f}")

env.close()
```

```python
from stable_baselines3 import DQN

env = gym.make("LunarLander-v3")

dqn_model = DQN(
    "MlpPolicy",
    env,
    verbose=1
)

dqn_model.learn(total_timesteps=100_000)

env.close()
```

**DQN training - summary**

At the beginning of training, the DQN agent achieved a very low average episode reward (around -270) and episodes were short, indicating frequent crashes. After 100,000 timesteps, the average reward improved to approximately -50 and episode length increased significantly.

**PPO training - summary**

At the start of training, the PPO agent had an average episode reward of about -144 with short episodes. By the end of training, the average reward increased to around -44 and episodes became much longer.

**Comparison**

Both algorithms achieved substantial improvements over the random agent, with PPO learning faster and reaching a higher final performance than DQN.

```python
import time
import gymnasium as gym
from stable_baselines3 import DQN, PPO

env = gym.make("LunarLander-v3")

start = time.time()
dqn_model = DQN("MlpPolicy", env, verbose=0)
dqn_model.learn(total_timesteps=50_000)
dqn_time = time.time() - start

start = time.time()
ppo_model = PPO("MlpPolicy", env, verbose=0)
ppo_model.learn(total_timesteps=50_000)
ppo_time = time.time() - start

env.close()

print("DQN training time:", dqn_time)
print("PPO training time:", ppo_time)
```
